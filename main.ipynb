```python
import re
def clean_html_tags(text):
    return re.sub(r'<[^>]+>.*?</[^>]+>', '', text)

df['Backstory'] = df['Backstory'].apply(clean_html_tags)
```

```python
invalid_chars_dict = {}

for col in df.columns:
        for text in df[col]:
            if isinstance(text, str):
                for char in text:
                    if not char.isalnum():
                        invalid_chars_dict[char] = True

unique_invalid_chars = list(invalid_chars_dict.keys())
print("Unique invalid characters:", unique_invalid_chars)
```

```python
for col in df.columns:
    df[col] = df[col].str.replace(u'\xa0', u' ')
    df[col] = df[col].str.replace(u'\u2003', u' ')
    df[col] = df[col].str.replace(u'\u200b', u' ')
    df[col] = df[col].str.replace(u'\u2028', u' ')
    df[col] = df[col].str.replace(u'\t', u' ')
```

```python
# Convert strings to lowercase
df.loc[:, 'Race'] = df['Race'].str.lower()
df.loc[:, 'Class'] = df['Class'].str.lower()
```

```python
raceList = ['human', 'astral elf', 'dwarf', 'half-elf', 'half-orc', 'halfling', 'tiefling', 'aasimar', 'orc', 'goliath', 'firbolg', 'kenku', 'tabaxi', 'dragonborn',
            'aarakocra', 'triton', 'genasi', 'tortle', 'changeling', 'kalashtar', 'shifter', 'warforged', 'centaur', 'loxodon', 'minotaur', 'simic hybrid', 'vedalken', 'verdan',
            'locathah', 'satyr', 'leonal', 'lizardfolk', 'grung', 'githzerai', 'githyanki', 'kobold', 'reborn', 'bugbear', 'hobgoblin', 'autognome', 'deep gnome', 'dhampir',
            'duergar', 'eladrin', 'fairy', 'giff', 'hadozee', 'harengon', 'hexblood', 'kender', 'leonin', 'owlin', 'plasmoid', 'sea elf', 'shadar-kai', 'thri-kreen',
            'yuan-ti pureblood', 'elf', 'gith', 'yuan-ti','gnome', 'goblin']

# Function to update race names
def update_race(race_name):
    for raceName in raceList:
        if raceName in race_name:
            return raceName
    return None

# Apply the function to the 'Class' column
df['Race'] = df['Race'].apply(update_race)

# Print out the unique values and their counts for the 'Class' column
df['Race'].value_counts()
```

```python
classList = ['artificer', 'barbarian', 'bard', 'cleric', 'druid', 'fighter', 'monk', 'paladin', 'ranger', 'rogue', 'sorcerer', 'warlock', 'wizard']
# Function to update class names
def update_class(class_name):
    for className in classList:
        if className in class_name:
            return className
    return None

# Apply the function to the 'Class' column
df['Class'] = df['Class'].apply(update_class)

#print out the unique values and their counts for the 'Class' column
df['Class'].value_counts()
```

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
OPENAI_API_KEY = os.getenv('OPENAI_API_KEY')

client = OpenAI(api_key=OPENAI_API_KEY)
```

```python
import json
with open('dnd_finetune.jsonl', 'w', encoding="utf-8") as file:
    for _, row in df.iterrows():
        response = client.chat.completions.create(model="gpt-3.5-turbo",
            messages=[
              {"role": "user", "content": f"I need a prompt. This prompt will be given to gpt and should ask him to write a backstory for a dnd character. And it should say that the character has the following information. Also the prompt should be short and as an answer just return the prompt and do not enclose it in quotation marks. The character's name is {row['Name']} and his class is {row['Class']} and he is of the {row['Race']} race."},
            ])
        json_obj = {
            "messages": [
                {"role": "system", "content": "You are an expert in crafting detailed backstories for Dungeons & Dragons characters. A user has provided you with basic character information: Name, Class, and Race. Your task is to create an extensive and engaging backstory. This backstory should reflect the character's history, psychological attributes, motivations, fears, and how their class attributes influence their life story. The backstory should be long and detailed, providing depth to the character. Remember, the goal is to enrich the user's gaming experience with a well-developed and coherent narrative."},
                {"role": "user", "content": response.choices[0].message.content},
                {"role": "assistant", "content": row['Backstory']}
            ]
        }
        file.write(json.dumps(json_obj, ensure_ascii=False) + '\n')
```

```python
response = client.files.create(
  file=open("dnd_finetune.jsonl", "rb"),
  purpose="fine-tune"
)
```
